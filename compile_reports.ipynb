```python
import shutil
import os
```

```python
DIRECTORY = 'results_1-250'
NEW_DIRECTORY = 'compiled_reports'
```

```python
subfolders = [f.name for f in os.scandir(DIRECTORY) if f.is_dir()]

print(subfolders)
```

```python
if not os.path.exists(NEW_DIRECTORY):
    os.makedirs(NEW_DIRECTORY)
```

```python
NEEDED_FILES = ['pre.json', 'report.json', 'file.json', 'registry.json']
first_subfolder = subfolders[0]
first_subfolder_path = os.path.join(DIRECTORY, first_subfolder)

nested_files_deep = []

for root, dirs, files in os.walk(first_subfolder_path):
    for file in files:
        if file in NEEDED_FILES:
            # nested_files_deep.append(file)
            nested_files_deep.append(os.path.join(root, file))

print(nested_files_deep)
```

```python
for subfolder in subfolders:
    skip_flag = False
    print(f"Process {subfolder}")

    src_path = os.path.join(DIRECTORY, subfolder)

    src_pre_file = os.path.join(src_path, 'pre.json')
    src_report_file = os.path.join(src_path, 'task_1/report.json')
    src_file_file = os.path.join(src_path, 'task_1/events/file.json')
    src_registry_file = os.path.join(src_path, 'task_1/events/registry.json')

    for src_file in [src_pre_file, src_report_file, src_file_file, src_registry_file]:
        if not os.path.exists(src_file):
            skip_flag = True
            print(f"{src_file} is missing.")
            print(f"{src_path} is invalid. Skipping.")
            break

    if skip_flag:
        continue

    dst_path = os.path.join(NEW_DIRECTORY, subfolder)
    if not os.path.exists(f'{dst_path}'):
        os.makedirs(f'{dst_path}')
    
    dst_pre_file = os.path.join(dst_path, 'pre.json')
    shutil.copy(src_pre_file, dst_pre_file)

    dst_report_file = os.path.join(dst_path, 'report.json')
    shutil.copy(src_report_file, dst_report_file)

    dst_file_file = os.path.join(dst_path, 'file.json')
    shutil.copy(src_file_file, dst_file_file)

    dst_registry_file = os.path.join(dst_path, 'registry.json')
    shutil.copy(src_registry_file, dst_registry_file)

    print(f"Processed {subfolder}.")
```
