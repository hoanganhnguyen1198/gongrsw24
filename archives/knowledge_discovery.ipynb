```python
import numpy as np
import pandas as pd
```

```python
PROCESSED_DS_DIR = "augusto_ds/processed_ds/"
```

```python
def get_family(df):
    family = []

    for i in range(len(df)):
        if 192<=df["id"][i]<=206:
            family.append("clop")
        elif(207<=df["id"][i]<=310):
            family.append("conti")
        elif(311<=df["id"][i]<=355):
            family.append("egregor")
        elif(403<=df["id"][i]<=435):
            family.append("lockbit")
        elif(438<=df["id"][i]<=451):
            family.append("lockbit")
        elif(452<=df["id"][i]<=468):
            family.append("mountlocker")
        elif(480<=df["id"][i]<=557):
            family.append("netwalker")
        elif(559<=df["id"][i]<=611):
            family.append("ryuk")
        elif(612<=df["id"][i]<=711):
            family.append("revil")
        elif(712<=df["id"][i]<=784):
            family.append("goodware")
        elif(791<df["id"][i]<1446):
            family.append("revil")
        else:
            family.append("unknown")

    # for i in range(len(family)):
    #     print(df['id'][i], family[i])
    
    return family
```

```python
df_1075 = pd.read_csv(PROCESSED_DS_DIR + '1075.csv', index_col=0)
df_1075
```

```python
family = get_family(df_1075)
```

```python
family_info = {}
for family_name in family:
    if family_name in family_info:
        family_info[family_name] += 1
    else:  
        family_info[family_name] = 1
family_info
```
